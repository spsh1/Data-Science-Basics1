# **Line Plot Basics**

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x=np.arange(1,11)
```

```python
print(x)
```

```python
y=2*x
```

```python
print(y)
```

```python
plt.plot(x,y)
plt.show()
```

## **Adding Titles and lables**

```python
plt.plot(x,y)
plt.title("Line Plot of x and y variable")
plt.xlabel("x-label")
plt.ylabel("y-label")
plt.show()
```

## **Changing Line Aesthetics**

```python
plt.plot(x,y,color="orange",linestyle=":",linewidth=3)
plt.title("Line Plot of x and y variable")
plt.xlabel("x-label")
plt.ylabel("y-label")
plt.show()
```

## **Adding two lines in the same plot**

```python
y1=2*x
y2=3*x
```

```python
plt.plot(x,y1,color="g",linewidth=3)
plt.plot(x,y2,color="blue",linestyle=":",linewidth=5)

plt.grid(True)
```

## **Adding subplots horizontally**

```python
plt.subplot(1,2,1)
plt.plot(x,y1,color="g",linewidth=3)

plt.subplot(1,2,2)
plt.plot(x,y1,color="r",linewidth=3)
plt.show()
```

## **Adding subplots vertically**

```python
plt.subplot(2,1,1)
plt.plot(x,y1,color="g",linewidth=3)

plt.subplot(2,1,2)
plt.plot(x,y1,color="r",linewidth=3)
plt.show()
```
