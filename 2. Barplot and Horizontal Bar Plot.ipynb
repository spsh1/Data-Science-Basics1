# **Bar Plot**

```python
import numpy as np
```

```python
from matplotlib import pyplot as plt
```

```python
student={"Aayush raj regmi":95, "Thomas Basyal":60,"Rishi":82}
```

```python
print(student)
```

```python
type(student)
```

```python
names=list(student.keys())
values=list(student.values())
```

```python
plt.bar(names,values)
plt.show()
```

## **Adding titles and lables to the bar plot**

```python
plt.bar(names,values)
plt.title("Bar Plot")
plt.xlabel("Name of a student")
plt.ylabel("Marks of a student")
plt.show()
```

## **Horizontal barplot**

```python
plt.barh(names,values, color="purple")
plt.title("Bar Plot")
plt.xlabel("Name of a student")
plt.ylabel("Marks of a student")
plt.grid(True)
plt.show()
```
