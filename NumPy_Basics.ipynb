# joining NumPy array

```python
import numpy as np
```

```python
a=np.array([10,20,30])
b=np.array([30,50,90])
```

```python
np.vstack((a,b))  #joining the numpy as vertical stacking 
```

```python
np.hstack((a,b)) #joining numpy as horizontal stacking 
```

```python
np.column_stack((a,b)) #joining numpy as column stacking
```

# NumPy intersection and difference

```python
import numpy as np
```

```python
x=np.array([10,20,30,40,50,60])
```

```python
y=np.array([30,50,90,100,120,150])
```

```python
np.intersect1d(x,y)
```

```python
np.setdiff1d(x,y)
```

# Addition of NumPy Arrays

```python
import numpy as np
```

```python
t=np.array([10,20])
s=np.array([30,40])
```

```python
np.sum([t,s])
```

```python
np.sum([t,s],axis=0)
```

```python
np.sum([t,s],axis=1)
```
