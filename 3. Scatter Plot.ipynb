# **Creating a Basic Scatter Plot**

```python
import matplotlib.pyplot as plt
```

```python
x=[10,20,30,40,50,60,70,80,90]
a=[8,1,7,2,0,3,7,3,2]
```

```python
plt.scatter(x,a)
plt.show()
```

## **Markers in the scatter plot**

```python
plt.scatter(x,a, marker="*",c="orange",s=50)
plt.show()
```

# **Adding two markers in the same plot**

```python
z=[10,20,30,40,50,60,70,80,90]
x=[8,1,7,2,0,3,7,3,2]
y=[7,2,5,6,9,1,4,5,3]
```

```python
plt.scatter(z,x, marker="*",c="g",s=100)
plt.scatter(z,y, marker=".",c="r",s=200)
plt.show()
```

## **Adding two subplots**

```python
z=[10,20,30,40,50,60,70,80,90]
x=[8,1,7,2,0,3,7,3,2]
y=[7,2,5,6,9,1,4,5,3]

plt.subplot(1,2,1)
plt.scatter(z,x, marker="*",c="g",s=100)

plt.subplot(1,2,2)
plt.scatter(z,y, marker=".",c="r",s=200)
plt.show()
```
